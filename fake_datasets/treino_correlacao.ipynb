```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt

n = 100

x = np.random.randint(150,190,n)/100
y = np.arange(0,n,1)
z = np.random.randint(20,28,n)

df = pd.DataFrame({'ind':y, 'altura':x})
df['peso'] = df.altura**2 * z
df['altura_desvio_media'] = df.altura - df.altura.mean()
df['altura_desvio_std'] = df.altura_desvio_media / df.altura.std()

df['peso_desvio_media'] = df.peso - df.peso.mean()
df['peso_desvio_std'] = df.peso_desvio_media / df.peso.std()
df = df.sort_values('altura')

df.peso_desvio_std.sort_values()
```

```python
fig, ax = plt.subplots(4,3, figsize = [16,16], )
fig.patch.set_facecolor('Grey')
lala = [(x,y) for x in range(0,4) for y in range(0,3)]

for x,y in lala:
    ax[x,y].set_facecolor((0, 0, 0))


df.plot(y = 'peso', kind = 'hist', ax = ax[0,0])
df.plot(y = 'altura', kind = 'hist', ax = ax[0,1])

df.plot(y = 'peso_desvio_media', kind = 'hist', ax = ax[1,0])
df.plot(y = 'altura_desvio_media', kind = 'hist', ax = ax[1,1])

df.plot(y = 'peso_desvio_std', kind = 'hist', ax = ax[2,0])
df.plot(y = 'altura_desvio_std', kind = 'hist', ax = ax[2,1])

df.plot(y = 'peso', x = 'altura', kind = 'scatter', ax = ax[3,0])
df.plot(y = 'peso_desvio_media', x = 'altura_desvio_media', kind = 'scatter', ax = ax[3,1])
df.plot(y = 'peso_desvio_std', x = 'altura_desvio_std', kind = 'scatter', ax = ax[3,2])
plt.tight_layout()
print(f'correlation: {(df.peso_desvio_std * df.altura_desvio_std).mean()}')
```
