```python
def s(n):
    import matplotlib.pyplot as plt
    x=[1]
    y=[0]
    c=0
    d=1
    for i in range(n):
        y.append(y[-1]+(n-c))
        y.append(y[-1])
        x.append(x[-1])
        x.append(x[-1]+n-d) #avan√ßa
        c+=1
        d+=1
        y.append(y[-1]-(n-c))
        y.append(y[-1])
        x.append(x[-1])
        x.append(x[-1]-n+d) #retrocede
        d=d+1
        c = c+1
    y[0]=1
    return x[0:2*n],y[0:2*n]
```

```python
import matplotlib.pyplot as plt, numpy as np
from matplotlib.patches import Ellipse

t=np.array(s(5))*3
fig,ax = plt.subplots(figsize=[4,4])
ax.plot(t[0],t[1],t[0]+40,t[1])
ax.add_artist(plt.Circle([20,0]*3,40, fill=None))

ax.plot([-x**2+30 for x in np.linspace(-2,2,10)],np.linspace(-5,5,10), color='k')
ax.plot([x**2-19 for x in np.linspace(-2,1,10)],np.linspace(-10,5,10), color='k')

ax.fill_between(np.linspace(-20,60,10),np.linspace(21,28,10),
                [-x**2 + x + 42 for x in np.linspace(-4,4,10)], color='k')

ax.add_patch(Ellipse((30,-20),20,10,facecolor='black'))

ax.set_yticks([-50,-20,0,20,50])
ax.set_xticks([-30,-20,0,30,70])
plt.grid()
```
