```python
import matplotlib.pyplot as plt, numpy as np
from matplotlib.patches import Circle
from matplotlib.patches import Ellipse
```

```python
fig = plt.figure(figsize=(16,9))
ax = fig.add_axes([0,0,1,1])

#head
head = Ellipse((2,5),1,3,fill=None)
l_eye = Circle((1.8,5),0.1,fill=None)
ax.plot(1.8,5,2.2,5, marker='.', color='k')
r_eye = Circle((2.2,5),0.1,fill=None)
ax.add_patch(head)
ax.add_patch(l_eye)
ax.add_patch(r_eye)

nose = ax.plot([2,2.01],[4.6,4.4], color='k')

mouth = ax.plot(np.linspace(-0.1,0.1,10) + 2,[x**2 +4 for x in np.linspace(-0.2,0.2,10)], color='k')

hair1 =np.linspace(-0.5,0.5,10)+2
hair2 = [-x**2 + 6.5 for x in np.linspace(-0.8,0.8,10)]
hair3 = np.linspace(5.7,5.7,10)
#ax.plot()
ax.fill_between(hair1,hair2,hair3, color='k', alpha=1)

ears = ax.plot([x**2 + 1.4 for x in np.linspace(-0.3,0.3,10)],np.linspace(4.8,5.2,10), color='k')
ears2 = ax.plot([-x**2 + 2.6 for x in np.linspace(-0.3,0.3,10)],np.linspace(4.8,5.2,10), color='k')

ax.set_yticks([2,4,5,6,8])
ax.set_xticks([0,2,4])
ax.grid()
```
