```python
import numpy as np, matplotlib.pyplot as plt

fig = plt.figure(figsize=[4,4])
ax = fig.add_axes([0,0,1,1])

#corpo
ax.plot([(x**2)/2 +  20.5 for x in np.linspace(-3.5,5,100)],np.linspace(-40,40,100), color='k')
ax.plot([-(x**2)/2 -  20.5 for x in np.linspace(-3.5,5,100)],np.linspace(-40,40,100), color='k')

#peitos
ax.plot(np.linspace(-25,-2,100),[x**2 +23 for x in np.linspace(-3.4,3.4,100)],color='k')
ax.plot(np.linspace(2,25,100),[x**2 +23 for x in np.linspace(-3.4,3.4,100)],color='k')
ax.plot()

#coxas
ax.plot([-(x**2)/2 + 2*x +  33  for x in np.linspace(-2,3,100)],np.linspace(-40,-80,100), color='k')
ax.plot([(x**2)/2 - 2*x -  33  for x in np.linspace(-2,3,100)],np.linspace(-40,-80,100), color='k')

#pernas
ax.plot(np.linspace(0,-3,100),np.linspace(-60,-100,100), color='k')
ax.plot(np.linspace(0,3,100),np.linspace(-60,-100,100), color='k')
ax.plot(np.linspace(0,-3,100),np.linspace(-60,-100,100), color='k')
ax.plot(np.linspace(35,32,100),np.linspace(-80,-100,100), color='k')
ax.plot(np.linspace(-35,-32,100),np.linspace(-80,-100,100), color='k')

#bra√ßos
ax.plot([-34,-20],[40,-20], color ='k')
ax.plot([-46,-25],[40,-40], color ='k')
ax.plot([34,20],[40,-20], color ='k')
ax.plot([46,25],[40,-40], color ='k')

#ombro
ax.plot(np.cos(np.linspace(0,1.5,100))*20+26,np.sin(np.linspace(0,1,100))*20+40, color='k')
ax.plot(-np.cos(np.linspace(0,1.5,100))*20-26,np.sin(np.linspace(0,1,100))*20+40, color='k')

#saco de papel
ax.fill_between(np.linspace(-30,30,68),np.sin(np.linspace(-34,34,68))*2+55,np.linspace(120,120,68),
                facecolor='sandybrown',
               edgecolor='black')
ax.plot([-30,-30],[55,120], color='black')
ax.plot([30,30],[55,120], color='black')
ax.plot([-30,30],[120,120], color='black')

#olhos
x = plt.Circle([-10,100],5, facecolor='k')
x2 = plt.Circle([10,100],5, facecolor='k')
ax.add_artist(x)
ax.add_artist(x2)

ax.set_xticks([-100,-44,-34,-6,6,34,100])
ax.set_yticks([-100,-80,-40,-60,-20,20,40,55,100])

plt.show()
```
