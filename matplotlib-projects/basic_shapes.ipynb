```python
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots(2,2)

#plot a straight horizontal line

ax[0,0].plot([0,1],[1,1], label='horizontal line')

#plot a straight vertical line
ax[0,1].plot([1,1],[0,1], label='vertical line')

#plot a positive diagnonal line
ax[1,0].plot([0,1],[0,1], label='positive \ndiagonal line')

#plot a negative diagnonal line
ax[1,1].plot([0,1],[1,0], label='negative \ndiagonal line')

for x in range(2):
    for y in range(2):
        ax[x,y].legend(loc=2, fontsize=8)
ax[1,1].legend(loc=1, fontsize=8)
```

```python
import numpy as np
plt.style.use('default')
fig,ax = plt.subplots(2,2)

#plot a horizontal u-shaped curve
ax[0,0].plot(np.arange(-1,1.1,0.1) + 2,[x**2 +1 for x in np.arange(-1,1.1,0.1)])
ax[0,0].set_title('horizontal u-shaped')

#plot a horizontal inverse u-shaped curve
ax[0,1].plot(np.arange(-1,1.1,0.1) + 2,[-x**2 +1 for x in np.arange(-1,1.1,0.1)])
ax[0,1].set_title('horizontal negative u-shaped')

#plot a vertical left-skewed curve
ax[1,0].plot([x**2+1 for x in np.arange(-1,1.1,0.1)], np.arange(-1,1.1,0.1))
ax[1,0].set_title('vertical left u-shaped')

#plot a vertical right-skewed curve
ax[1,1].plot([-x**2+1 for x in np.arange(-1,1.1,0.1)], np.arange(-1,1.1,0.1))
ax[1,1].set_title('vertical right u-shaped')
plt.tight_layout()
```

```python
#different ways of plotting a circle
fig, ax = plt.subplots(2,2,sharey=True,sharex=True)
# the hard way
o = 1
a = np.arange(1000,-1000,-0.1)
h = np.sqrt(a**2 + o**2)

s = np.concatenate((o/h,-o/h))
c = np.concatenate((a/h,-a/h))
ax[0,0].plot(s,c)

# a little better
ax[0,1].plot(np.sin(np.arange(-3.1,3.2,0.1)),np.cos(np.arange(-3.1,3.2,0.1)))

# way better
xxx = plt.Circle((0,0),1,edgecolor='blue',fill=None)
ax[1,0].add_artist(xxx)

# ideal
from matplotlib.patches import Circle
ax[1,1].add_patch(Circle((0,0),1,fill=None))
```

```python
#ellipsis
fig, ax = plt.subplots(1,3,figsize=(8,4))
# the hard way
o = 1
a = np.linspace(1000,-1000,10000)
h = np.sqrt(a**2 + o**2)

s = np.concatenate((o/(2*h),-o/(2*h)))
c = np.concatenate((a/h,-a/h))
ax[0].plot((s),(c),color='k')
ax[0].set_xticks([-1,1])

#ideal
from matplotlib.patches import Ellipse
ax[1].add_patch(Ellipse((0.5,0.5),0.5,1,fill=None))

plt.tight_layout()

#tilted
from matplotlib.patches import Ellipse
ax[2].add_patch(Ellipse((0.5,0.5),0.5,1,fill=None, angle=45))

plt.tight_layout()
```
